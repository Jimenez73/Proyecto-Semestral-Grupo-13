```python
import clase_scraper
```

```python
# Funciones útiles:

def dataframe_columna_dividida(df):
    posicion_original = df.columns.get_loc("Won - Lost")
    columnas_creadas = df.pop("Won - Lost").str.split(" - ", expand=True)
    columnas_creadas.columns = ["Won", "Lost"]
    for i, col in enumerate(columnas_creadas.columns):
        df.insert(posicion_original + i, col, columnas_creadas[col])

    return df
```

```python
hltv = clase_scraper.HltvScraper()
```

```python
# Terroristas

df_tt = hltv.pistol_rounds(side = "tt")
```

```python
df_tt
```

```python
df_tt_dividido = dataframe_columna_dividida(df_tt)
df_tt_dividido
```

```python
# Ambos: 

df_both = hltv.pistol_rounds(side = "both")
```

```python
df_both
```

```python
df_both_dividido = dataframe_columna_dividida(df_both)
df_both_dividido
```

```python
# Contraterroristas: 

df_ct = hltv.pistol_rounds(side = "ct")
```

```python
df_ct
```

```python
df_ct_dividido = dataframe_columna_dividida(df_ct)
df_ct_dividido
```
